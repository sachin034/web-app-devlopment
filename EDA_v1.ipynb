### For each type of auction, how many were held in each template and how has this varied over time?

#### Auction Type = Auction

```python
all_data_auction=all_data.loc[all_data['EventType']=='Auction']
all_data_auction['ExtendedPriceAmountInReportingCurrency']=all_data_auction['ExtendedPriceAmountInReportingCurrency'].replace('NA',0)
all_data_auction['ExtendedPriceAmountInReportingCurrency']=all_data_auction['ExtendedPriceAmountInReportingCurrency'].replace(' ',0)
all_data_auction['ExtendedPriceAmountInReportingCurrency']=all_data_auction['ExtendedPriceAmountInReportingCurrency'].apply(pd.to_numeric, errors='coerce')
all_data_auction.pivot_table(index='Template', columns='Year', values=['DocId'], aggfunc='count', fill_value=0)
```

```python
all_data_auction.pivot_table(index='Template', columns='Year', values=['ExtendedPriceAmountInReportingCurrency'], aggfunc='sum', fill_value=0)
```

#### Findings for auction type - 'Auction' :
#### 1.  Popular templates for auction
        - 'Modified reverse auction' (72% of total extended price)
        - 'modified reverse auction bid transformation' (24% of total extended price)
        - 'English_Reverse_eTender_Self_Serve' has been introduced only this year with growing popularity (3% of total extended price)
